```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['seaborn-darkgrid'])

import numpy.random as rnd
```

---
## Q1 &mdash; All candidates

What is the probability that Barbara gets the best candidate if she interviews all of the candidates? 

 * Use small numbers so not drowned by too much data.
 * Fixed random seed so that results are reproducible (for now).
 * Don't worry about speed - just get it correct.
 * Print statements are your friend - insert as many as you need to see what is happening

```python
def trial(N):
    interviews = rnd.permutation(range(N))
    return interviews[-1]==0

def run_trials(n, N):
    return sum(trial(N) for _ in range(n)) / n

N_values = [1,2,4,8,16,32]
pr_values = [run_trials(10_000, N) for N in N_values]
```

```python
plt.plot(N_values, pr_values)
plt.xlabel("Number of candidates, $N$")
plt.suptitle("Probability of selecting best candidates")
plt.title("'Interview all candidates' strategy")
plt.ylabel("Probability")
plt.show()
```

---
## Q2 &mdash; First candidate

What is the probability that Barbara gets the best candidate if she chooses the first candidate?

 * Graphics are useful when moving to larger data values.
 * For graphs, we usually need to keep record of history of data - use lists/arrays.
 * Compare output with earlier iterations - use fixed seeds.

```python
def trial(N):
    interviews = rnd.permutation(range(N))
    return interviews[0]==0

def run_trials(n, N):
    return sum(trial(N) for _ in range(n)) / n

N_values = [1,2,4,8,16,32]
pr_values = [run_trials(10_000, N) for N in N_values]
```

```python
plt.plot(N_values, pr_values)
plt.xlabel("Number of candidates, $N$")
plt.suptitle("Probability of selecting best candidates")
plt.title("'Select first candidate' strategy")
plt.ylabel("Probability")
plt.show()
```
